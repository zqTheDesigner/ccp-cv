```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img_path = './cat.jpg'
```

```python
img = cv.imread(img_path)
```

```python
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
img.shape
```

```python
img2 = cv.resize(img,
                 None,
                 fx=0.5, 
                 fy=0.5, 
                 interpolation=cv.INTER_LINEAR
                )
```

```python
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
```

```python
img3 = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
cv.imshow('image', img3)

# waits for user to press any key
# (this is necessary to avoid Python kernel form crashing)
cv.waitKey(0)
  
# closing all open windows

cv.destroyAllWindows()

cv.waitKey(1)
```

```python
plt.imshow(cv.cvtColor(img3, cv.COLOR_BGR2RGB))
```

```python
print(img.shape)
print(img2.shape)
print(img3.shape)
```

```python
img4 = cv.GaussianBlur(img3, (21, 21), 0)
```

```python
plt.imshow(cv.cvtColor(img4, cv.COLOR_BGR2RGB))
```

```python
img5 = cv.Canny(img4, 40, 40)
```

```python
plt.imshow(cv.cvtColor(img5, cv.COLOR_BGR2RGB))
```

```python
cv.imshow('image', img)
cv.imshow('resize', img2)
cv.imshow('gray', img3)
cv.imshow('blur', img4)
cv.imshow('Canny Edge Detection', img5)

# waits for user to press any key
# (this is necessary to avoid Python kernel form crashing)
cv.waitKey(0)
  
# closing all open windows

cv.destroyAllWindows()

cv.waitKey(1)
```
